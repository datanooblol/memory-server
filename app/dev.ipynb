```python
from package.data_models import User, Project, ChatSession, Source, SourceType, Conversation, Role, Reference, ReferenceType, AgentExecution
from package.repositories.factory import RepositoryFactory
from package.storage.factory import StorageFactory
import package.schemas as schemas

config = dict(path="./data/test.db")

user_storage = StorageFactory.create_relational("duckdb", config)
project_storage = StorageFactory.create_relational("duckdb", config)
chat_session_storage = StorageFactory.create_relational("duckdb", config)
source_storage = StorageFactory.create_relational("duckdb", config)
conversation_storage = StorageFactory.create_relational("duckdb", config)
reference_storage = StorageFactory.create_relational("duckdb", config)
agent_execution_storage = StorageFactory.create_relational("duckdb", config)
```

```python
await user_storage.create_table("users", schemas.USER)
await project_storage.create_table("projects", schemas.PROJECT)
await chat_session_storage.create_table("chat_sessions", schemas.CHAT_SESSION)
await source_storage.create_table("sources", schemas.SOURCE)
await conversation_storage.create_table("conversations", schemas.CONVERSATION)
await reference_storage.create_table("references", schemas.REFERENCE)
await agent_execution_storage.create_table("agent_executions", schemas.AGENT_EXECUTION)
```

```python
user_repo = RepositoryFactory(user_storage).get_user_repository()
project_repo = RepositoryFactory(project_storage).get_project_repository()
chat_session_repo = RepositoryFactory(chat_session_storage).get_chat_session_repository()
source_repo = RepositoryFactory(source_storage).get_source_repository()
conversation_repo = RepositoryFactory(conversation_storage).get_conversation_repository()
reference_repo = RepositoryFactory(reference_storage).get_reference_repository()
agent_execution_repo = RepositoryFactory(agent_execution_storage).get_agent_execution_repository()
```

```python
user = User(email="user@test.com", password="string")
user_id = await user_repo.create(user)
user_id
```

```python
project = Project(user_id=user_id, project_name="Test", project_description="This is project description.")
project_id = await project_repo.create(project)
project_id
```

```python
source = Source(project_id=project_id, source_name="tips.csv", source_type=SourceType.CSV, source_path=dict(path="./tips.csv"))
source_id = await source_repo.create(source)
source_id
```

```python
chat_session = ChatSession(project_id=project_id)
chat_session_id = await chat_session_repo.create(chat_session)
chat_session_id
```

```python
user_convo = Conversation(chat_session_id=chat_session_id, role=Role.USER, content="Hi")
user_convo_id = await conversation_repo.create(user_convo)
user_convo_id
```

```python
ai_convo = Conversation(chat_session_id=chat_session_id, role=Role.ASSISTANT, content="How are you?")
ai_convo_id = await conversation_repo.create(ai_convo)
ai_convo_id
```

```python
for t, c in [
    (ReferenceType.SQL_CODE, "sql code"),
    (ReferenceType.SQL_DATA, "sql data"),
    (ReferenceType.PLOTLY_CODE, "plotly code"),
    (ReferenceType.PLOTLY_DATA, "plotly data"),
]:
    reference = Reference(convo_id=ai_convo_id, type=t, content=c)
    reference_id = await reference_repo.create(reference)
# reference_id
```

```python
await reference_repo.find_by(dict(convo_id=ai_convo_id))
```

```python
first_created_reference = await reference_repo.get_first_n(1)
first_created_reference = first_created_reference[0]
first_created_reference
```

```python
updated_reference = Reference(convo_id=ai_convo_id, type=ReferenceType.SQL_CODE, content="sql code updated")
await reference_repo.update(first_created_reference.reference_id, updated_reference)
await reference_repo.get_by_id(first_created_reference.reference_id)
```
