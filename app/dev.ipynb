```python
import duckdb
from pathlib import Path
def query_csv_in_duckdb(files:list[str], query):
    conn = duckdb.connect()
    for file in files:
        table_name = Path(file).stem  # Gets "tips" from "tips.csv"
        conn.execute(f"CREATE VIEW {table_name} AS SELECT * FROM '{file}'")
    # data = conn.execute(query).fetchall()
    # fields = [d[0] for d in conn.description]
    # return fields, data
    data = conn.execute(query).df()
    conn.close()
    return data.to_markdown()
```

```python
data = query_csv_in_duckdb(files=['./project_dataset/2f8499d5-9d23-44f6-ae3f-18c547a7341e/tips.csv'], query="SELECT * FROM tips LIMIT 5;")
print(data)
```

```python
def to_toon(fields, data):
    lines = [",".join(fields)]
    for d in data:
        lines.append(",".join(d))
    return "\n".join(lines)
```
